```python
from copy import deepcopy
class puzzle_8():
    def __init__(self, informacion, similitud = None):
        self.__informacion = informacion
        self.__similitud = similitud
        self.__hijos = None
        self.__padre = None

    #Getters y setters
    @property
    def informacion(self):
        return self.__informacion 

    @informacion.setter 
    def informacion(self, nueva_informacion):
        self.__informacion = nueva_informacion

    @property
    def similitud(self):
        return self.__similitud

    @similitud.setter
    def similitud(self, nueva_similitud):
        self.__similitud = nueva_similitud

    @property 
    def hijos(self):
        return self.__hijos

    @hijos.setter 
    def hijos(self, nuevos_hijos):
        self.__hijos = nuevos_hijos

        for cada_hijo in nuevos_hijos:
            cada_hijo.__padre = self 

    @property 
    def padre(self):
        return self.__padre

    @padre.setter
    def padre(self, nuevo_padre):
        self.__padre = padre

    #Funciones de la clase puzzle
    def comparar_con_solucion(self, solucion):
        casillas_coincidentes = 0
        puzzle = self.informacion

        for fila in range(len(puzzle)):
            for columna in range(len(puzzle[fila])):
                if puzzle[fila][columna] == solucion[fila][columna] and puzzle[fila][columna] != None:
                    casillas_coincidentes += 1

        self.similitud = casillas_coincidentes

    def encontrar_hueco(self):
        for n_fila, fila in enumerate(self.informacion):
            if None in fila: 
                posicion_hueco = (n_fila, fila.index(None))

        return posicion_hueco

    def encontrar_vecinos(self, posicion):
        '''
        posicion = (x, y) -> x == fila; y == columna
        '''
        fila = posicion[0]
        columna = posicion[1]

        longitud_fila = len(self.informacion[fila])
        longitud_columna = len(self.informacion)

        if fila == 0: 
            if columna == 0:
                vecinos = [(fila + 1, columna), (fila, columna + 1)]

            elif columna == longitud_columna - 1:
                vecinos = [(fila + 1, columna), (fila, columna - 1)]

            else:
                vecinos = [(fila, columna + 1), (fila, columna - 1), (fila + 1, columna)]

        elif fila == longitud_fila - 1:
            if columna == 0:
                vecinos = [(fila - 1, columna), (fila, columna + 1)]

            elif columna == longitud_columna - 1:
                vecinos = [(fila - 1, columna), (fila, columna - 1)]

            else: 
                vecinos = [(fila - 1, columna), (fila, columna - 1), (fila, columna + 1)]

        else: 
            if columna == 0:
                vecinos = [(fila, columna + 1), (fila - 1, columna), (fila + 1, columna)]

            elif columna == longitud_columna - 1:
                vecinos = [(fila, columna - 1), (fila - 1, columna), (fila + 1, columna)]

            else:
                vecinos = [(fila - 1, columna), (fila + 1, columna), (fila, columna - 1), (fila, columna + 1)]

        valores_vecinos = []
        for x, y in vecinos:
            valores_vecinos.append(self.informacion[x][y])

        return vecinos

    #Representación

    def __str__(self):
        matriz = self.informacion[:]

        return '''
        {}
        {}
        {}
        Similitud: {}'''.format(matriz[0], matriz[1], matriz[2], self.similitud)
```

```python
def solucionar(puzzle_inicial, puzzle_final):
    puzzle_actual = puzzle_8(puzzle_inicial)
    puzzle_actual.comparar_con_solucion(puzzle_final)

    posibles_movimientos = []
    posibles_movimientos.append(puzzle_actual)

    movimientos_realizados = []
    movimientos_realizados.append(puzzle_actual.informacion)

    while len(posibles_movimientos) != 0:
        posibles_movimientos.sort(key = lambda puzzle: puzzle.similitud, reverse = True)

        puzzle_actual = posibles_movimientos[0]
        posibles_movimientos.pop(0)

        if puzzle_actual.informacion == puzzle_final:
            return puzzle_actual

        else:
            posicion_hueco = puzzle_actual.encontrar_hueco()
            posicion_vecinos = puzzle_actual.encontrar_vecinos(posicion_hueco)

            fila_hueco = posicion_hueco[0]
            columna_hueco = posicion_hueco[1]

            lista_hijos = []
            for x, y in posicion_vecinos:
                un_hijo = deepcopy(puzzle_actual.informacion)

                '''
                print("Son el mismo objeto?: ", un_hijo is puzzle_actual.informacion)
                print("Un hijo antes del cambio: \t", un_hijo)
                print("Puzzle actual antes del cambio: ", puzzle_actual.informacion)
                '''

                valor_vecino = un_hijo[x][y]
                un_hijo[x][y] = None
                un_hijo[fila_hueco][columna_hueco] = valor_vecino

                '''
                print("Un hijo después del cambio: \t", un_hijo)
                print("puzzle actual despues del cambio: ", puzzle_actual.informacion, end = "\n\n")
                '''
                puzzle_hijo = puzzle_8(un_hijo)
                puzzle_hijo.comparar_con_solucion(puzzle_final)

                lista_hijos.append(puzzle_hijo)


            puzzle_actual.hijos = lista_hijos

            for hijo in puzzle_actual.hijos:
                if hijo.informacion not in movimientos_realizados:
                    movimientos_realizados.append(hijo.informacion)
                    posibles_movimientos.append(hijo)


def mostrar_resolucion(puzzle, puzzle_inicial):
    lista_pasos = []
    while puzzle.informacion != puzzle_inicial:
        lista_pasos.append(puzzle)
        puzzle = puzzle.padre
    lista_pasos.append(puzzle)
    lista_pasos.reverse()
    
    for num_paso, paso in enumerate(lista_pasos):
        print("Paso : {} \n {}\n".format(num_paso, paso))
        
```

```python
solucion_puzzle = [
    [1, 2, 3],
    [8, None, 4],
    [7, 6, 5]
]

puzzle_inicial = [
    [2, 8, 3],
    [1, 6, 4],
    [7, None, 5]
]

puzzle_solucionado = solucionar(puzzle_inicial, solucion_puzzle)
mostrar_resolucion(puzzle_solucionado, puzzle_inicial)

print("---------Nuevo puzzle-----------")

puzzle_inicial_2 = [
    [4, 5, 8],
    [None, 1, 3],
    [7, 6, 2]
]

puzzle2_solucionado = solucionar(puzzle_inicial_2, solucion_puzzle)
mostrar_resolucion(puzzle2_solucionado, puzzle_inicial_2)
```
